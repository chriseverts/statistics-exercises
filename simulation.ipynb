```python
%matplotlib inline
import numpy as np
import pandas as pd


np.random.seed(29)



##1. How likely is it that you roll doubles when rolling two dice?

n_trials = nrows = 10_000
n_dice = ncols = 2

roll1 = np.random.choice([1,2,3,4,5,6], n_trials * n_dice).reshape(nrows,ncols)
roll2 = np.random.choice([1,2,3,4,5,6], n_trials * n_dice).reshape(nrows,ncols)
doubles = (roll1 == roll2).mean()

doubles
print(f'The probability to roll doubles is {doubles}')


```

```python
##2. If you flip 8 coins, what is the probability of getting exactly 3 heads? 
##What is the probability of getting more than 3 heads?

n_trials = nrows = 10_000
n_coins = ncols = 8
result = np.random.choice(['heads', 'tails'], n_trials * n_coins).reshape(nrows,ncols)# Calculates the posibilities
heads = (result == 'heads').sum(axis =1)# extracts the heads possibilities


three_heads = (heads ==3).mean()# calculates the possibilities of 3 heads
three_heads

```

```python
n_trials = nrows = 10_000
n_coins = ncols = 8
result = np.random.choice(['heads', 'tails'], n_trials * n_coins).reshape(nrows,ncols)
heads = (result == 'heads').sum(axis =1)


greater_heads = (heads >3).mean()# calculates the possibilities of more then 3 heads
greater_heads
```

```python
##3. There are approximitely 3 web development cohorts for every 1 data science cohort at Codeup.
##Assuming that Codeup randomly selects an alumni to put on a billboard, what are the odds that the two billboards I drive past both have data science students on them?

n_trials = nrows = 10_000
n_billb = ncols = 2
result = np.random.choice(['Data Science', 'WD'], n_trials * n_billb).reshape(nrows,ncols)## selects random options from a list (Data science or Web dev)
dsresults = (result == 'Data Science').sum(axis =1)
ds = (dsresults ==2).mean()## takes the result of "data science" that will go on both billboards

ds
```

```python
##4. Codeup students buy, on average, 3 poptart packages (+- 1.5) a day from the snack vending machine. 
## If on monday the machine is restocked with 17 poptart packages, how likely is it that I will be able to buy some poptarts on Friday afternoon?
poptart_avg = 3 ## average of 3 pop tarts a day
poptart_stdev = 1.5 ## standard deviation of 1.5 a day
n_trials = nrows = 10_000
n_days = ncols = 5

pt_bought = np.random.normal(poptart_avg,poptart_stdev,size=(n_trials,n_days)) ## generates numbers from a normal distribution with a specified mean and standard deviation
pt_bought

```

```python
pt_on_friday = 17 - pt_bought.sum(axis =1)## subtracts total number from restock on monday
pt_on_friday
```

```python
(pt_on_friday >0).mean()
```

```python
##5. Compare Heights

##Men have an average height of 178 cm and standard deviation of 8cm.
##Women have a mean of 170, sd = 6cm.
##Since you have means and standard deviations, you can use np.random.normal to generate observations.
##If a man and woman are chosen at random, P(woman taller than man)?


mheight_avg = 178
mheight_stdev = 8

wheight_avg = 170
wheight_stdev = 6

n_trials = nrows = 10_000
n_subj = ncols = 2

mstats = np.random.normal(mheight_avg,mheight_stdev,size=(n_trials,n_subj)) ## generates numbers from a normal distribution with a specified mean and standard deviation
mstats

wstats = np.random.normal(wheight_avg,wheight_stdev,size=(n_trials,n_subj)) ## generates numbers from a normal distribution with a specified mean and standard deviation
wstats

wtaller = (wstats > mstats).mean()## checks to see at random woman taller then man
wtaller
```

```python
##If 23 people are in the same room, what are the odds that two of them share a birthday? What if it's 20 people? 40?


data = range(0,365)
ntrials = 23
nsimulations = 10_000

rooms = np.random.choice(data, size= (nsimulations, ntrials))
rooms
```

```python
list_double_bdays = [len(np.unique(rooms[n])) for n in range(0, nsimulations-1) if len(np.unique(rooms[n])) < 23]
list_double_bdays
```

```python
same_bdays = len(list_double_bdays) / nsimulations
same_bdays
```
